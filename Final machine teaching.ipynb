```python
# работы 1
# товары 2
# услуги 3

# 2    265473
# 3     50347
# 1     24379

import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.externals import joblib

vse = pd.read_csv('ste_cats.csv',usecols=['Наменование','cat'])


vse.append({'Наменование' : 'Инкассация' , 'cat' : 3} , ignore_index=True)


cat_y = vse['cat'].values
train = vse['Наменование'].astype('str')


vectorizer = TfidfVectorizer(analyzer='char_wb', ngram_range=(2, 3), max_features=4000)
vectorizer.fit(train)

filename = 'vectorizer.pkl'
_ = joblib.dump(vectorizer, filename, compress=9)


X = vectorizer.transform(train)



from sklearn.model_selection import cross_val_score
from sklearn import linear_model

from sklearn.utils import shuffle
X,Y = shuffle(X,cat_y)


#lg = linear_model.RidgeClassifier()

lg = linear_model.LogisticRegression()


lg.fit(X,Y)

print(cross_val_score(lg,X,Y, cv=3, scoring="accuracy", verbose=3))


filename = 'vid_prod.joblib.pkl'
_ = joblib.dump(lg, filename, compress=9)


print("dump saved!")

cont1 =  pd.read_excel('cont_2_1_inns.xlsx', usecols = ['Предмет'])
#display(cont1.head(1))

cont2 =  pd.read_excel('cont_2_2_inns.xlsx', usecols = ['Предмет'])

cont1=cont1.append(cont2)

P = vectorizer.transform(cont1['Предмет'].astype('str'))

print(P.shape)
cont1['Вид продукции'] = lg.predict(P)
cont1.to_csv('предмет_вид_продукции_1')

cont1.head(30)
```
